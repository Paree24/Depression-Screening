```python
def genderclean(x):
    x=x.strip()
    x=x.lower()
    if "female" in x:
        return 0
    if "male" in x:
        return 1
    else:
        return 2
    return x
data=pd.read_csv("cleaned.csv")
if data.columns[0]=='Unnamed: 0':
    data=data.iloc[:,1:]
```

```python
data_to_encode=data.select_dtypes(include=['object'])
data_to_encode.columns
```

```python
from category_encoders import TargetEncoder
```

```python
for col in data_to_encode.columns:
    data[col]=TargetEncoder().fit_transform(X=data[col],y=data['Depression'])
```

```python
y=data['Depression']
```

```python
y
```

```python
x=data.drop('Depression',axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=512,random_state=35).fit(xtrain,ytrain)
```

```python
pred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier(random_state=12).fit(xtrain,ytrain)
pred=model.predict(xtest)
from sklearn.metrics import accuracy_score
accuracy_score(ytest,pred)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model=AdaBoostClassifier(random_state=12).fit(xtrain,ytrain)
pred=model.predict(xtest)
from sklearn.metrics import accuracy_score
accuracy_score(ytest,pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(random_state=22).fit(xtrain,ytrain)
pred=model.predict(xtest)
from sklearn.metrics import accuracy_score
accuracy_score(ytest,pred)
```
